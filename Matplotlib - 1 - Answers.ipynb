1. Draw a line with suitable label in the x axis, y axis and a title

```python
X = range(1, 50)
Y = [v * 3 for v in X]


plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('line')

plt.show()
```

3. Plot 2 lines on same plot with suitable legends of each line <br>
#### hint ( plt.legend() )

```python
x1 = [10,20,30]
y1 = [20,40,10]


x2 = [10,20,30]
y2 = [40,10,30]



plt.plot(x1, y1)

plt.plot(x2, y2)

plt.xlabel('x - axis')

plt.ylabel('y - axis')

plt.title('2 lines with suitable legends ')

plt.legend(['f line', 's line'])

plt.show()
```

```python
x1 = [10,20,30]
y1 = [20,40,10]


x2 = [10,20,30]
y2 = [40,10,30]



plt.plot(x1, y1, label = 'f line')

plt.plot(x2, y2,  label = 's line')

plt.xlabel('x - axis')

plt.ylabel('y - axis')

plt.title('2 lines with suitable legends ')

plt.legend()

plt.show()
```

4. Plot 2 lines with legends, different widths and colors

```python
x1 = [10,20,30]
y1 = [20,40,10]

x2 = [10,20,30]
y2 = [40,10,30]

plt.xlabel('x - axis')

plt.ylabel('y - axis')

plt.title('2 lines with different widths and colors')

plt.plot(x1,y1, color='green', linewidth = 3,  label = 'line1-width-3')
plt.plot(x2,y2, color='yellow', linewidth = 5,  label = 'line2-width-5')

plt.legend()
plt.show()
```

5. Plot 2 lines with different styles

```python
x1 = [10,20,30]
y1 = [20,40,10]

x2 = [10,20,30]
y2 = [40,10,30]

plt.xlabel('x - axis')

plt.ylabel('y - axis')

plt.plot(x1,y1, color='green', linewidth = 3,  label = 'line1-dotted',linestyle='dotted')
plt.plot(x2,y2, color='yellow', linewidth = 5,  label = 'line2-dashed', linestyle='dashed')

plt.title("2 lines with different styles")

plt.legend()

plt.show()
```

6. Draw a line and set the line markers

```python

x = [1,4,5,6,7]

y = [2,6,3,6,3]


plt.plot(x, y, color='blue', linestyle='dashdot', linewidth = 3,
         marker='o', markerfacecolor='yellow', markersize=12)


#Set the y-limits of the current axes.
plt.ylim(1,8)
#Set the x-limits of the current axes.
plt.xlim(1,8)

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('Marker')

plt.show()
```

7. Display the current axis limits values and set new axis values
#### hint ( plt.axis() )

```python
X = range(1, 50)
Y = [value * 3 for value in X]


plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('line')

# shows the current axis limits values
print(plt.axis()) 

# Limit of x axis -1.4 to 51.4
# Limit of y axis -4.2 to 154.2

plt.axis([-5,60,-20,160])


print(plt.axis()) 


plt.show()
```

9. Plot 3 lines with different format styles in one command using arrays

```python
import numpy as np
import matplotlib.pyplot as plt

# Sampled time at 100ms intervals
t = np.arange(0., 5., 0.2)
print(t)

# green dashes, blue squares and red triangles
plt.plot(t, t, 'bs--', t, t**2, 'gs', t, t**3, 'y^')
plt.show()
```
